```python
a = float(input("Введите a: "))
b = float(input("Введите b: "))
eps = float(input("Введите eps: "))

if a > b:
    a, b = b, a

def fun(x):
    return x**4 - 4*x**3 - 8*x**2 + 1

def find_roots(a, b, eps):
    roots = []
    step = (b - a) / 100

    for i in range(100):
        x1 = a + i * step
        x2 = a + (i + 1) * step

        f1 = fun(x1)
        f2 = fun(x2)

        if f1 * f2 < 0:
            while (x2 - x1) >= eps:
                c = (x1 + x2) / 2
                fc = fun(c)
                if f1 * fc < 0:
                    x2 = c
                    f2 = fc
                else:
                    x1 = c
                    f1 = fc
            roots.append(c)

    return roots

roots = find_roots(a, b, eps)

if roots:
    print("Найденные корни:", roots[0])
else:
    print("Нет корней на выбранном отрезке")
```
