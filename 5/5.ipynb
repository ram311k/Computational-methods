```python
import numpy as np

def f(x):
    return np.sin(x**2)

def trapezoidal_rule(a, b, tol):
    n = 1
    integral_old = 0.0
    k = 0
    
    while True:
        h = (b - a) / n
        integral_new = 0.5 * (f(a) + f(b))

        for i in range(1, n):
            integral_new += f(a + i * h)

        integral_new *= h
        
        if abs(integral_new - integral_old) < tol:
            break
        
        integral_old = integral_new
        n *= 2
        k += 1
        
    print(k)

    return integral_new

a = 0
b = np.pi / 4
tolerance = 0.0001

result = trapezoidal_rule(a, b, tolerance)
print(result)
```
